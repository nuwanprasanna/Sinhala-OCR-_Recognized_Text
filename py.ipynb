```python
import cv2
import pytesseract
from PIL import Image
import matplotlib.pyplot as plt

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def extract_text_from_image(image_path):
    # Read the image using OpenCV
    img = cv2.imread(image_path)

    # Check if the image is successfully loaded
    if img is None:
        print("Error: Could not read the image.")
        return

    # Get the image size (height, width)
    height, width = img.shape[:2]

    # Convert the image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Use pytesseract to extract text
    extracted_text = pytesseract.image_to_string(Image.fromarray(gray_img))

    # Split the text into lines
    lines = extracted_text.split('\n')

    # Display the original image
    plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
    plt.imshow(gray_img)
    plt.title(f'Original Image - Size: {width}x{height} pixels')
    plt.axis('off')
    
    plt.show()

    # Display the extracted text line by line
    print("Extracted Text:")
    for line in lines:
        print(line)

# Provide the path to your input image
input_image_path = './img.jpeg'

# Call the function to extract text and display the result
extract_text_from_image(input_image_path)
```
